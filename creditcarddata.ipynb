```python
df['is_fraud'] = df['is_fraud'].str.extract('(\d+)').astype('float').astype('Int64')
```

```python
df['is_fraud'] = df['is_fraud'].astype('int')
fraud_by_merchant = df.groupby('merchant')['is_fraud'].mean().sort_values(ascending=False)
fraud_by_merchant
```

**Transaction Overview**

1.The overall distribution of transaction amounts.

```python
sns.histplot(df['amt'], bins=50, kde=True)
plt.title('Histogram of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(df['amt'])
plt.title('Box Plot of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.show()

mean_amt = df['amt'].mean()
median_amt = df['amt'].median()
std_amt = df['amt'].std()
print("Mean:", mean_amt)
print("Median:", median_amt)
print("Standard:", std_amt)
```

*The histogram of transaction amounts shows a right-skewed distribution with a long tail to the right. This means that there are a few very large transaction amounts that pull the mean to the right, while the majority of transactions are smaller.*


*The box plot suggests that most transactions are relatively small, with a few outliers on the high end. The IQR is relatively small, indicating that the majority of transactions are clustered around the median value.*

2.There are specific states or cities where transaction amounts are higher.

```python
avg_amt_by_state = df.groupby('state')['amt'].mean().sort_values(ascending=False)

sns.barplot(x=avg_amt_by_state.index, y=avg_amt_by_state)
plt.title('Average Transaction Amount by State')
plt.xlabel('State')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=45)
plt.show()
```

*The bar plot titled "Average Transaction Amount by State" shows the average transaction amount for each of the 12 states listed on the x-axis. The y-axis represents the average transaction amount, ranging from 0 to 250.*

3.The proportion of fraudulent transactions in the dataset

```python
fraud_counts = df['is_fraud'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(fraud_counts, labels=fraud_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fraudulent Transactions')
plt.show()
```

*The pie chart illustrates the distribution of fraudulent transactions in a dataset. It shows the proportion of transactions that are labeled as fraudulent (1) compared to those labeled as non-fraudulent (0).*

**Temporal Analysis**

1.The rate of fraudulent transactions per hour.

```python
df['date_time'] = pd.to_datetime(df['trans_date_trans_time'], format='%d-%m-%Y %H:%M')
df['hour'] = df['date_time'].dt.hour
fraudulent_hours = df.groupby('hour')['is_fraud'].mean().sort_values(ascending=False)
fraudulent_hours.plot(kind='bar')
plt.title('Rate of fraudulent transactions per Hour')
plt.ylabel('Rate')
plt.xlabel('Hour (24hours)')
```

X-axis Label Clarity: We adjusted the x-axis label to "Hour (24-hour format)" for better clarity.
Horizontal X-axis Labels: We set plt.xticks(rotation=0) to keep the hour labels horizontal, as rotating them wouldn't be as readable in this case.

*The bars with the highest values indicate the hours when fraudulent transactions are most likely to occur. the hours with lower bars represent periods when fraud is less prevalent.the hours with lower bars represent periods when fraud is less prevalent.*

2.The count of fraudulent transactions per month

```python
df['month'] = df['date_time'].dt.month_name()
fraudulent_months = df.groupby('month')['is_fraud'].sum().sort_values(ascending=False)
fraudulent_months.plot(kind='bar')
plt.title('Count of fraudulent transactions per month')
plt.ylabel('Count')
plt.xlabel('Month')
```

Descriptive Y-axis Label: We changed the y-axis label to "Count of Fraudulent Transactions" for clarity.
X-axis Label Rotation: We rotated the x-axis labels by 45 degrees for better readability, as month names can be long.

* The months with the tallest bars experienced the highest number of fraudulent transactions. This could indicate seasonal patterns in fraud activity.By understanding the monthly distribution of fraud, you can proactively adjust fraud prevention strategies to address potential risks throughout the year.*

**Demographical Analysis**

1.Fraudulent data among customers of certain age groups

```python
df['dob'] = pd.to_datetime(df['dob'], format='%d-%m-%Y')
df['yob'] = df['dob'].dt.year
df['age'] = 2019-df['yob']
bins = [0, 24, 39, 54, 74, float('inf')]
labels = ['Gen Z', 'Millennials', 'Gen X', 'Baby Boomers', 'Silent Generation']
df['agegroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)
```

```python
sns.barplot(data=df, x='agegroup', y='is_fraud', palette='viridis')
plt.title('Fraudulent Transactions by Age group')
plt.xticks(rotation=45)
plt.xlabel('Age Group')
plt.ylabel('Rate')
```

Accurate Y-axis Label: We changed the y-axis label to "Fraud Rate" to clearly indicate that you're plotting the proportion of fraudulent transactions within each age group, not the total count.

*The age groups with the tallest bars have the highest fraud rates, suggesting they are more susceptible to fraudulent activity. Understanding the relationship between age and fraud can help businesses and financial institutions develop more targeted fraud prevention strategies for specific age groups.*

This project involved conducting an exploratory data analysis (EDA) on a credit card transaction dataset to uncover patterns and characteristics associated with fraudulent activity. The analysis revealed a right-skewed distribution of transaction amounts, with high-value outliers and an overall fraud rate of approximately 12.77%. Temporal patterns indicate that fraud is more common at certain times of the day and during specific months, while geographic analysis shows significant variation in fraud rates across states and cities. Categorical analysis identified transaction categories and job types with higher fraud rates, and demographic analysis highlighted susceptibility by age group. There is a strong correlation between transaction amounts and fraud was observed. The dataset also exhibited class imbalance, with far more non-fraudulent transactions than fraudulent ones. These insights are crucial for refining fraud detection systems, improving risk management, and implementing targeted prevention strategies.
